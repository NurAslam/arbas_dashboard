```python
import pandas as pd
import os
```

```python
# Tentukan path file
base_path = '/Users/user/Documents/03 KERJA/PT Multimedia Solusi Prima/2026/APRIL/Arbas Market Intelligence - Data/Dashboardanlys/'

# Load customer data
df_customers = pd.read_csv(os.path.join(base_path, 'customers.csv'))

# Tampilkan info dasar
print(f"✅ Total Records: {len(df_customers):,}")
print(f"\n📋 Columns: {list(df_customers.columns)}")
print(f"\n🔍 Sample Data:")
df_customers
```

```python
# Konversi createdAt ke datetime
df_customers['createdAt'] = pd.to_datetime(df_customers['createdAt'], format='mixed')

# Tentukan tanggal referensi (hari ini)
today = pd.Timestamp.today().normalize()
days_ago_7  = today - pd.Timedelta(days=7)
days_ago_30 = today - pd.Timedelta(days=30)
days_ago_90 = today - pd.Timedelta(days=90)

# Filter new user per periode
new_7d  = df_customers[df_customers['createdAt'] >= days_ago_7]
new_30d = df_customers[df_customers['createdAt'] >= days_ago_30]
new_90d = df_customers[df_customers['createdAt'] >= days_ago_90]

print(f"📅 Referensi tanggal hari ini: {today.date()}")
print("=" * 50)
print(f"🟢 NEW (7 hari terakhir):  {len(new_7d):>5} records")
print(f"🔵 NEW (30 hari terakhir): {len(new_30d):>5} records")
print(f"🟣 NEW (90 hari terakhir): {len(new_90d):>5} records")
print("=" * 50)
```

```python
# ============================================================
# ANALISIS CUSTOMER BARU — 30 HARI TERAKHIR
# Fokus: Kota, Tipe Channel, Ukuran Outlet
# ============================================================

print(f"📅 Customer Baru — 30 Hari Terakhir ({days_ago_30.date()} s/d {today.date()})")
print(f"   Total customer baru: {len(new_30d)} records\n")

# --- Kota ---
print("🏙️  DISTRIBUSI KOTA")
print("-" * 45)
kota_counts = new_30d['kota'].value_counts(dropna=False).reset_index()
kota_counts.columns = ['kota', 'jumlah']
kota_counts['persen'] = (kota_counts['jumlah'] / len(new_30d) * 100).round(1)
print(kota_counts.to_string(index=False))

# --- Tipe Channel ---
print("\n📡 DISTRIBUSI TIPE CHANNEL")
print("-" * 45)
channel_counts = new_30d['tipeChannel'].value_counts(dropna=False).reset_index()
channel_counts.columns = ['tipeChannel', 'jumlah']
channel_counts['persen'] = (channel_counts['jumlah'] / len(new_30d) * 100).round(1)
print(channel_counts.to_string(index=False))

# --- Ukuran Outlet ---
print("\n🏪 DISTRIBUSI UKURAN OUTLET")
print("-" * 45)
outlet_counts = new_30d['ukuranOutlet'].value_counts(dropna=False).reset_index()
outlet_counts.columns = ['ukuranOutlet', 'jumlah']
outlet_counts['persen'] = (outlet_counts['jumlah'] / len(new_30d) * 100).round(1)
print(outlet_counts.to_string(index=False))
```

```python
# Detail data customer baru 30 hari — kolom pilihan
print("\n🔍 DETAIL DATA — Customer Baru 30 Hari Terakhir")
print("   (Kota | Tipe Channel | Ukuran Outlet)\n")
new_30d[['customerCode', 'nama', 'kota', 'tipeChannel', 'ukuranOutlet', 'provinsi', 'status', 'createdAt']]\
    .sort_values('createdAt', ascending=False)\
    .reset_index(drop=True)
```

```python
# ============================================================
# BREAKDOWN tipeChannel — SELURUH DATA df_customers
# ============================================================

print("📡 DISTRIBUSI tipeChannel (SEMUA DATA)")
print(f"   Total records: {len(df_customers)}\n")

channel_all = df_customers['tipeChannel'].value_counts(dropna=False).reset_index()
channel_all.columns = ['tipeChannel', 'jumlah']
channel_all['persen'] = (channel_all['jumlah'] / len(df_customers) * 100).round(1)
print(channel_all.to_string(index=False))

print("\n🔍 SAMPLE data per tipeChannel:")
for channel in df_customers['tipeChannel'].dropna().unique():
    subset = df_customers[df_customers['tipeChannel'] == channel]
    print(f"\n  [{channel}] — {len(subset)} records")
    display(subset[['customerCode', 'nama', 'kota', 'ukuranOutlet', 'status']].head(3))
```

```python
import matplotlib.pyplot as plt

# ============================================================
# VISUALISASI — DISTRIBUSI tipeChannel (SELURUH DATA)
# ============================================================

# Siapkan data
channel_df = df_customers['tipeChannel'].value_counts(dropna=False).reset_index()
channel_df.columns = ['tipeChannel', 'jumlah']
# Ganti NaN label jadi 'Unknown'
channel_df['tipeChannel'] = channel_df['tipeChannel'].fillna('(kosong)')

# Set warna
colors = plt.cm.Set2(range(len(channel_df)))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Bar Chart ---
ax1 = axes[0]
bars = ax1.barh(channel_df['tipeChannel'], channel_df['jumlah'], color=colors)
ax1.set_xlabel('Jumlah Customer')
ax1.set_ylabel('tipeChannel')
ax1.set_title('Distribusi tipeChannel (Bar)', fontsize=13, fontweight='bold')
for bar, val in zip(bars, channel_df['jumlah']):
    ax1.text(val + 5, bar.get_y() + bar.get_height()/2, str(val), va='center', fontsize=10)
ax1.set_xlim(0, channel_df['jumlah'].max() * 1.15)

# --- Pie Chart ---
ax2 = axes[1]
wedges, texts, autotexts = ax2.pie(
    channel_df['jumlah'],
    labels=channel_df['tipeChannel'],
    autopct='%1.1f%%',
    colors=colors,
    startangle=140
)
ax2.set_title('Distribusi tipeChannel (Pie)', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()

# ============================================================
# VISUALISASI — CUSTOMER BARU 7 / 30 / 90 HARI
# ============================================================

periods = {'7 Hari': new_7d, '30 Hari': new_30d, '90 Hari': new_90d}

fig2, axes2 = plt.subplots(1, 3, figsize=(18, 5))

for ax, (label, df_period) in zip(axes2, periods.items()):
    channel_period = df_period['tipeChannel'].value_counts(dropna=False).reset_index()
    channel_period.columns = ['tipeChannel', 'jumlah']
    channel_period['tipeChannel'] = channel_period['tipeChannel'].fillna('(kosong)')

    bars2 = ax.bar(channel_period['tipeChannel'], channel_period['jumlah'], color=plt.cm.Set3(range(len(channel_period))))
    ax.set_title(f'tipeChannel — {label}\n({len(df_period)} records)', fontsize=11, fontweight='bold')
    ax.set_xlabel('tipeChannel')
    ax.set_ylabel('Jumlah')
    ax.tick_params(axis='x', rotation=30)
    for b, v in zip(bars2, channel_period['jumlah']):
        ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.5, str(v), ha='center', fontsize=9)

plt.suptitle('Distribusi tipeChannel — Customer Baru per Periode', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
